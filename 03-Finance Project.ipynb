# Finance EDA 



```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

## Data

We need to get data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo


```python
import datetime
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2020,1,1)
BAC = data.DataReader("BAC", 'yahoo', start, end)
BAC

```

```python
citi=data.DataReader("C", 'yahoo', start, end)
citi
```

```python
goldman=data.DataReader("GS", 'yahoo', start, end)
goldman
```

```python
jp=data.DataReader("JPM", 'yahoo', start, end)
jp
```

```python
morgan=data.DataReader("MS", 'yahoo', start, end)
morgan
```

```python
fargo=data.DataReader("WFC", 'yahoo', start, end)
fargo
```

```python
tickers=['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

Concatenating

```python
bank_stocks=pd.concat([BAC,citi,goldman,jp,morgan,fargo],keys=tickers,axis=1)
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA


What is the max Close price for each bank's stock throughout the time period?

```python
bank_stocks.xs('Close',axis=1,level='Stock Info').max()
```

```python
returns=pd.DataFrame()
```

We use pct_change() method on the Close column to create a column representing this return value

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
import seaborn as sns
sns.pairplot(returns[1:])
```

what dates each bank stock had the best and worst single day returns?


```python
returns.idxmin()
```

```python
returns.idxmax()
```

 which stock would you classify as the riskiest over the entire time period?

```python
returns.std()
```

 Which would you classify as the riskiest for the year 2019?

```python
returns.loc['2019-01-01':'2019-12-31'].std() # Very similar risk profiles
```

2019 returns for Morgan Stanley

```python
sns.distplot(returns.loc['2019-01-01':'2019-12-31']['MS Return'],bins=100,color='green')
```

2019 returns for CitiGroup

```python
sns.distplot(returns.loc['2019-01-01':'2019-12-31']['C Return'],bins=100,color='red')
```
